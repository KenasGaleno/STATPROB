```python
# Fungsi untuk menghitung variansi
def calculate_variance(data):
    # Rata-rata
    mean = sum(data) / len(data)
    print(f"Rata-rata: {mean}")

    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")
    squared_diff = [(x - mean) ** 2 for x in data]
    for i, x in enumerate(data):
        print(f"({x} - {mean})**2 = {squared_diff[i]}")

    # Variansi
    variance = sum(squared_diff) / len(data)
    print(f"Variansi: {variance}")

# Inisialisasi dataset
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data}")

# Memanggil fungsi untuk menghitung variansi
calculate_variance(data)
```

```python
# Fungsi untuk menghitung rata-rata
def calculate_mean(data):
    return sum(data) / len(data)

# Fungsi untuk menghitung variansi
def calculate_variance(data):
    # Rata-rata
    mean = calculate_mean(data)
    print(f"Rata-rata: {mean}")

    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    squared_diff = []
    for x in data:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append(squared_diff_per_x)

    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata
    variance = sum(squared_diff)

    # Variansi: jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data - 1
    variance = variance / (len(data) - 1)

    print(f"\nVariansi: {variance}")

    return variance

# Fungsi untuk menghitung standar deviasi
def calculate_standard_deviation(data):
    variance = calculate_variance(data)
    standard_deviation = variance**0.5
    print(f"\nStandard Deviation: {variance}**0.5 = {standard_deviation}")

    return standard_deviation

# Inisialisasi dataset
data = [180, 185, 175, 190, 170, 195, 165]
print(f"Data: {data}")

# Hitung rata-rata dan standar deviasi
mean = calculate_mean(data)
standard_deviation = calculate_standard_deviation(data)

# Visualisasi data dengan histogram dan garis vertikal pada standar deviasi
import matplotlib.pyplot as plt
import numpy as np

plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean:.2f}')
plt.axvline(x=mean - standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-standard_deviation:.2f}')
plt.axvline(x=mean + standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean+standard_deviation:.2f}')

plt.xlabel('Tinggi (cm)')
plt.ylabel('Jumlah')
plt.title(f'Data Tinggi Badan | Std Deviation = {standard_deviation:.2f}')
plt.legend()
plt.show()
```
